```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

raw_sms = pd.read_csv('/content/spam.csv',encoding='latin-1')
print(raw_sms)
sms_data = raw_sms.where((pd.notnull(raw_sms)),'')
sms_data.head()
sms_data.loc[sms_data['v1'] == 'spam', 'v1'] = 0
sms_data.loc[sms_data['v1'] == 'ham', 'v1'] = 1
X = sms_data['v2']

Y = sms_data['v1']
print(X)
print(Y)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=4)

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)



Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
print(X_train)
print(X_train_features)

algorthim = LogisticRegression()
algorthim.fit(X_train_features, Y_train)

prediction_on_training_data = algorthim.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data : ', accuracy_on_training_data)
prediction_on_test_data = algorthim.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data : ', accuracy_on_test_data)

input_sms=["Nah I don't think he goes to usf, he lives around here though"]
input_data_features = feature_extraction.transform(input_sms)
prediction = algorthim.predict(input_data_features)
print(prediction)
if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
